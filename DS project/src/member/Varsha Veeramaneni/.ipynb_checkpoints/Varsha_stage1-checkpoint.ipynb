Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading Confirmed cases data sets files

```python
no_of_cases_file =("../../../data/stage_1/covid_confirmed_usafacts.csv")
no_of_cases = pd.read_csv(no_of_cases_file)
no_of_cases_df = pd.DataFrame(no_of_cases)
no_of_cases_df
```

Getting Last week data from Confirmed cases dataset

```python
lastweek= no_of_cases_df.iloc[: , -7:]
lastweek
```

Plotting graph to know the last weeks trends of Covid

```python
lastweektrends = lastweek.sum(axis=0)
graph = lastweektrends.plot(grid=True)
graph.set_ylabel('No of Cases in the United States')
graph.set_xlabel('Last week covid cases Trend')
graph.tick_params(axis='x', labelrotation = 50)
```

As we can see in the above trend the last weeks total confirmed cases is raising exponentially. From 8/10 to 8/12 and 8/15 to 8/16 there is high increase of covid confirmed cases. On the other days it has increased steadily.

Fetching Virginia state's last week data

```python
va_state = no_of_cases_df[no_of_cases_df['State'] == "VA"]
va_lastweek = va_state.iloc[: , -7:]
va_lastweek
```

Plotting graph to know the trends of Virginia state last week covid data

```python
vatrends = va_lastweek.sum(axis=0)
graph = vatrends.plot(grid=True)
graph.set_ylabel('No of Cases in the state of Virginia')
graph.set_xlabel('Last week covid cases Trend')
graph.tick_params(axis='x', labelrotation = 50)
```

As we can see in the above trend the last weeks total confirmed cases is raising exponentially. From 8/10 to 8/12 and 8/15 to 8/16 there is high increase of covid confirmed cases. On the other days there was a constant raise.

### Merging Enrichment dataset files based on common variables

```python
merged_dataset_new=merged_dataset.copy()
merged_dataset_new['County Name_x'] = merged_dataset_new['County Name_x'].apply(lambda cn: cn.strip())
merged_dataset_new
```
