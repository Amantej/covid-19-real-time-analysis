# Generate weekly statistics (mean, median, mode) for number of new cases and deaths across a specific state.

Here I'm using the large_covid data set to perform this task

```python
mergedcases_perday= pd.read_csv('/../../../data/stage_1/long_large_covid.csv')
```

```python
mergedcases_perday
```

I'm changing the date format

```python
mergedcases_perday['Date'] = mergedcases_perday.Date.apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
```

# Covid-19 new cases analysis:

Here I'm choosing 'FL' state to calculate weekly Statistics for the new covid -19 cases,deaths per day.

```python
FL_state=mergedcases_perday[mergedcases_perday['State']=='FL']
FL_state
```

To get individual day cases,deaths I'm using diff function and making column wise difference and storing these values in two new colums 

```python
FL_state['Num of Cases Per Day'] = FL_state['Cases'].diff()
FL_state['Num of Deaths Per Day'] = FL_state['Death'].diff()
```

```python
num = FL_state['Num of Cases Per Day']._get_numeric_data()
num[num < 0] = 0
num1 = FL_state['Num of Deaths Per Day']._get_numeric_data()
num1[num1 < 0] = 0
```

```python
FL_state
```

removing first date 2020-01-22 as it gives NaN

```python
FL_state = FL_state[FL_state['Date'] != '2020-01-22']
FL_state
```

```python
FL_state
```

Removing the number of cases and deaths columns as we have 2 new coulms with Num of cases and deaths per day

```python
 FL_state.drop(['Cases','Death'], axis=1, inplace=True)
```

here we are taking sum of number of cases,deaths and population and then gropingby state 

```python
f={'Num of Cases Per Day':'sum','Num of Deaths Per Day':'sum','Population':'sum','State':'first'}
FLstate_date = FL_state.groupby("Date").agg(f).reset_index()
FLstate_date
```

# Here we are normalizing(normalizing by 100000) the Florida state number of cases and deaths per day and storing in 2 new columns

```python

FLstate_date['Num of Cases Per Day Normalized']= (FLstate_date['Num of Cases Per Day']/FLstate_date['Population'])*100000
FLstate_date['Num of Deaths Per Day Normalized']= (FLstate_date['Num of Deaths Per Day']/FLstate_date['Population'])*100000

FLstate_date
```

After normalizing we are calculating the mean for the cases and the deaths

```python

FLstate_data = FLstate_date.groupby([ pd.Grouper(key='Date', freq='w')])['Num of Cases Per Day Normalized','Num of Deaths Per Day Normalized'].agg({'Num of Cases Per Day Normalized':'mean',
                                                                                                'Num of Deaths Per Day Normalized' :'mean'}).round().reset_index()
FLstate_data
```

For the "FL" state I'm calculating mean,median and mode for the normalized number of cases per day

```python

FLstate_data["Num of Cases Per Day Normalized"].agg(["mean","median"])
```

For the "FL" state I'm calculating mean,median for the normalized number of deaths per day

```python

FLstate_data["Num of Deaths Per Day Normalized"].agg(["mean","median"])
```

For the "FL" state I'm calculating mode for the normalized number of cases per day

```python
FLstate_data["Num of Cases Per Day Normalized"].mode()[0]
```

For the "FL" state I'm calculating mode for the normalized number of deaths per day

```python
FLstate_data["Num of Deaths Per Day Normalized"].mode()[0]
```

# Compare the data against other states (compare against 5 states)

```python

mergedcases_perday = mergedcases_perday.groupby(['State',"Date"]).sum().reset_index()
```

```python
mergedcases_perday
```

here we are grouping the states with population

```python
mergedcases_perday.groupby(['State','Population']).size().reset_index().rename(columns={0:'count'}).sort_values('Population')
```

Here I'm choosing 5 different states along with my Initial choosen state to make the easy comparison in the further steps

```python
states_covid_data = mergedcases_perday[mergedcases_perday["State"].isin(["FL","NC", "TX", "CA", "VA", "OH",])]
states_covid_data
```

To get individual day cases here I'm using diff function and making column wise difference

```python

states_covid_data['Num of Cases Per Day'] = states_covid_data['Cases'].diff()
states_covid_data['Num of Deaths Per Day'] = states_covid_data['Death'].diff()
```

```python
num = states_covid_data['Num of Cases Per Day']._get_numeric_data()
num[num < 0] = 0
num1 = states_covid_data['Num of Deaths Per Day']._get_numeric_data()
num1[num1 < 0] = 0
```

```python
states_covid_data = states_covid_data[states_covid_data['Date'] != '2020-01-22']
states_covid_data
```

Here I'm dropping the cases and deaths columns as we have new cases and deaths per day column

```python
states_covid_data.drop(['Cases','Death'], axis=1, inplace=True)
states_covid_data
```

Here I'm grouping the states by date

```python

f={'Num of Cases Per Day':'sum','Num of Deaths Per Day':'sum','Population':'sum'}
states_covid_data = states_covid_data.groupby(["State","Date"]).agg(f).reset_index()
states_covid_data
```

```python
states_covid_data
```

Here I'm normalizing the covid cases and deaths of all the selected states (normalizing the daily cases by 1000000)

```python

states_covid_data['Num of Cases Per Day Normalized']= (states_covid_data['Num of Cases Per Day']/states_covid_data['Population'])*100000
states_covid_data['Num of Deaths Per Day Normalized']= (states_covid_data['Num of Deaths Per Day']/states_covid_data['Population'])*100000
states_covid_data
```

```python
Meanof_5states = states_covid_data.groupby(["State", pd.Grouper(key='Date', freq='w')])['Num of Cases Per Day Normalized','Num of Deaths Per Day Normalized'].agg({'Num of Cases Per Day Normalized':'mean',
                                                                                                'Num of Deaths Per Day Normalized' :'mean',}).round().reset_index()
Meanof_5states
```

 calculating the mean and the median of cases for all the 6 states

```python

Meanof_5states.groupby('State')['Num of Cases Per Day Normalized'].agg(['mean','median']).round().reset_index()
```

calculating the mode of cases for all the 6 states

```python

Meanof_5states.groupby('State')['Num of Cases Per Day Normalized'].agg(lambda x:x.value_counts().index[0]).reset_index()
```

**From the above statistics **
- Though the population in California is higher when comapred with Florida and Texas,the weekly mean is less.
- Population in Florida is less when compared with Texas but the covid contagious rate is more in FL than in TX.
- For the 2nd and 3rd most populated states in US, FL and TX they do have similar stats, Florida has a higher weekly mean of cases that means it has higher cases than Texas in total.
- The population Varginia is less and the covid weekly cases and deaths mean is also less.s
