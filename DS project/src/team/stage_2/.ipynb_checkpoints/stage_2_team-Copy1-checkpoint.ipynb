### 1. Compare the weekly statistics (mean, median, mode) for number of new cases and deaths across US.

 ## Cases Statistics

Reading Covid cases dataset to understand weekly stats.

```python
covid_cases = pd.read_csv('../../../data/stage_2/covid_confirmed_usafacts.csv')
covid_cases.head()
```

Removing state rows as we are working on counties.

```python
covid_cases = covid_cases[covid_cases['countyFIPS'] != 0]
covid_cases.head()
```

Removing first four columns as they do not contribute to our requirement. And taking sum, row wise to understand total cummilative cases count on a given date.

```python
filter_data = covid_cases.iloc[:, 4:].sum()
filter_data_1 = pd.DataFrame(columns=covid_cases.columns[4:])
filter_data_1 = filter_data_1.append(filter_data, ignore_index=True)
filter_data_1
```

To get individual day cases, we are using diff function and making column wise difference and removing first date as it gives NaN

```python
filter_data_1 = filter_data_1.diff(axis=1)
filter_data_1.drop(columns=['2020-01-22'], inplace=True)
filter_data_1
```

Taking rounded total weeks value generated by given cases data and coverting given date, cases into weekly cases count. Also, appended column names as week starting date. 

```python
l = []
for i in range(len(filter_data_1.columns)//7+1): l.append(filter_data_1.iloc[:,i*7:i*7+7].sum(axis=1).astype(int)) 
```

```python
cases_weekly = pd.concat(l, axis=1)
cases_weekly = cases_weekly.set_axis(filter_data_1.columns[:(len(filter_data_1.columns)//7+1)*7:7],axis=1)
cases_weekly
```

Cal. Mean for individual weeks (sum of cases) 

```python
round(cases_weekly.mean(axis=1)).astype(int)
```

Cal. Median for individual weeks (sum of cases) 

```python
round(cases_weekly.median(axis=1)).astype(int)
```

Cal. Mode for individual weeks (sum of cases) 

```python
round(cases_weekly.mode(axis=1)).astype(int)
```

### 3. Compare the data against other countries of the world

## Covid data across the world

```python
world_data = pd.read_csv('../../../data/stage_2/owid-covid-data.csv')
world_data.head()
```

Filling NaN values with 0 as most of the columns are not considered for analysis. Also, selecting 5 countries which are close to United states population by reading/understanding data.

```python
world_data = world_data.fillna(0)
world_data_covid = world_data[world_data["location"].isin(["Indonesia", "Pakistan", "Brazil", "Nigeria", "Bangladesh",])]
```

```python
world_data_covid
```

Selecting columns which are exactly used for analysis or which matches with original US datasets.

```python
world_data_covid = world_data_covid[["location","date",'new_cases','new_deaths',"new_cases_per_million","new_deaths_per_million","population"]]
```

Applying datetime strptime to date conversion.

```python
from datetime import datetime
world_data_covid['date'] = world_data_covid.date.apply(lambda day: datetime.strptime(day, "%Y-%m-%d"))
```

### Start dates of each selected countries differ. We observed that every country has records from earliest date as 03-03-2020 and end date is matched with US original dataset timeline. 

```python
world_data_covid = world_data_covid[(world_data_covid['date'] > '2020-03-03') & (world_data_covid['date'] <= '2021-08-12')]
world_data_covid_weekly = world_data_covid.groupby('location').resample('W-Thu', on='date').mean().reset_index().sort_values(by=['location','date'])
```

```python
world_data_covid_weekly
```

United states weekly cases (sum of cases)

```python
cases_weekly_mean
```

Converting us weekly cases dataset to long format and restricting weeks which matches to world dataset

```python
us_new = pd.melt(cases_weekly_mean, var_name = 'date', value_name = 'new_cases')
us_new_1 = us_new[(us_new['date'] >= '2020-03-05') & (us_new['date'] <= '2021-08-16')]
us_new_1.head()
```

Changing data frame for readability

```python
deaths_weekly_mean
```

Applying the melt function on the above dataframe on date and selecting the data in between the dates matching with the world dataset.

```python
us_new1 = pd.melt(deaths_weekly_mean, var_name = 'date', value_name = 'new_deaths')
us_new_2 = us_new1[(us_new1['date'] >= '2020-03-05') & (us_new1['date'] <= '2021-08-16')]
us_new_2.head()
```

Adding new_deaths and location columns to the us_data dataframe

```python
us_data = us_new_1.copy()
us_data['new_deaths'] = us_new_2['new_deaths']
us_data['location'] = 'United States'
us_data.head()
```

Reading large covid dataset which we obtained from the stage_1. Calculating the sum of the entire Us population. Adding population, new_cases_per_million, new_deaths_per_million columns to the us_data dataframe. Changing the data type of date in us_dataframe to datetime.

```python
population_us = pd.read_csv('../../../data/stage_2/large_covid.csv')
us_population = population_us.population.sum()
us_data['population'] = us_population
us_data['new_cases_per_million']= abs((us_data['new_cases']/us_data['population'])*1000000)
us_data['new_deaths_per_million'] = abs((us_data['new_deaths']/us_data['population'])*1000000)
us_data['date'] = us_data.date.apply(lambda day: datetime.strptime(day, "%Y-%m-%d"))
us_data.reset_index(drop=True,inplace =True)
us_data.head()
```

```python
#saving dataset for stage 05
us_data_copy = us_data.copy()
import numpy as np
us_data_copy['Log normalized cases']= np.log(us_data_copy['new_cases_per_million'])
us_data_copy['Log normalized deaths']= np.log(us_data_copy['new_deaths_per_million'])
us_data_copy['Log normalized cases'].clip(lower=0,inplace = True)
us_data_copy['Log normalized deaths'].clip(lower=0,inplace = True)
us_data_copy
us_data_copy.to_csv('../../../data/stage_2/norm_usa.csv')
```

Appending us_data dataframe to the world_data_covid_weekly dataframe and store it in combined_data dataframe

```python
combined_data = world_data_covid_weekly.append(us_data)
combined_data.reset_index(drop = True, inplace = True)
```

```python
combined_data
```

Calculating the mean, median of new_cases grouping by location

```python
combined_data1 = combined_data.groupby('location')['new_cases'].agg(['mean','median']).reset_index().round()
combined_data1
```

From the above calculations we can observe that mean and median of new_cases is more in United States followed by Brazil. The Nigeria has the least mean and median.

Calculating the mean, median of new_deaths grouping by location

```python
combined_data2 = combined_data.groupby('location')['new_deaths'].agg(['mean','median']).reset_index().round()
combined_data2
```

Observations:
- Brazil has highest median and followed by United States.
We can say United States has a less death rate compared to Brazil.

Calculating the mean, median of new_cases_per_million grouping by location

```python
combined_data3 = combined_data.groupby('location')['new_cases_per_million'].agg(['mean','median']).reset_index().round()
combined_data3
```

Calculating the mean, median of new_deaths_per_million grouping by location

```python
combined_data4 = combined_data.groupby('location')['new_deaths_per_million'].agg(['mean','median']).reset_index().round()
combined_data4
```

```python
fig = pgo.Figure()
fig.add_trace(pgo.Bar(
    x=combined_data3['location'],
    y=combined_data3['mean'],
    name='mean',
    marker_color='blue'
))
fig.add_trace(pgo.Bar(
    x=combined_data3['location'],
    y=combined_data3['median'],
    name='median',
    marker_color='red'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45,
    title="Mean and median of the normalized number of cases across countries")

fig.write_image('../../../Images/team/Stage_2/fig.png')
f = fig.to_image(format="png")
Image(f)
```

### Try to do background research to see what the reasoning for the peaks can be (holidays, vacations, travel, etc). Explain and relate the peaks with a description in the notebooks.


- United States: United state first noted case in the end of January and reached highest peak after Christmas Holidays. Their is a downfall after April 2021 as vaccination process started. But again cases started increasing after July 2021 due to the new Variant.

- Brazil: Due to without proper covid measures the cases over brazil tend to incline. This resulted in burden on the brazilian hospital system throughout the covid waves resulted in high death rate after April 2021.

- Bangladesh: Bangladesh's socioeconomic and cultural conditions are unfavorable to maintain the guidelines given by WHO. And also the poor health care system of Bangladesh couldn't handle the spread of covid cases which resulted in increase in death rates after july 2021. Also vaccination process is delayed in this country compared to other countries.

- Pakistan: Due to lack of maintaining proper social distancing and also due to poor health care system the cases increased gradually Irrespective of the vaccination status.

- Indonesia: From the above plots we can see that there is a surge starting from June 2021 due to post-Ramdan ritual known as Mudik which resulted in severe spread of covid cases. Another reason apart from this is slower vaccination process compared to other countries.

- Nigeria: Due to lack of awareness and poor economic support in some regions of Nigeria resulted in spread of covid cases. And also due to lack of proper medicare and delayed vaccination process these covid trends continued to grow.



### 5. Identify the peak weeks of the cases and deaths in US and other countries

Finding weeks where maximum cases per million have registered grouping by location

```python
max_cases_weeks = combined_data.groupby(['location'])['new_cases_per_million'].transform(max) == combined_data['new_cases_per_million']
world_max_cases_week = combined_data[max_cases_weeks]
world_max_cases_week
```

Finding weeks where maximum deaths per million have registered grouping by location

```python
max_deaths_weeks = combined_data.groupby(['location'])['new_deaths_per_million'].transform(max) == combined_data['new_deaths_per_million']
world_max_deaths_week = combined_data[max_deaths_weeks]
world_max_deaths_week
```

```python
fig4 = pgo.Figure()
fig4.add_trace(pgo.Bar(
    x=world_max_cases_week['location'],
    y=world_max_cases_week['new_cases_per_million'],
    name='peak week cases',
    marker_color='blue'
))
fig4.add_trace(pgo.Bar(
    x=world_max_deaths_week['location'],
    y=world_max_deaths_week['new_deaths_per_million'],
    name='peak week deaths',
    marker_color='red'
))


fig4.update_layout(barmode='group', xaxis_tickangle=-45,
    title="Peak week of the cases and deaths in US and other countries")
fig4.write_image('../../../Images/team/Stage_2/fig4.png')
f4 = fig4.to_image(format="png")
Image(f4)
```
