Importing required libraries for dataframe manipulation and visual data representation.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

Reading the confirmed covid cases dataset.

```python
covid_confirmed_dataset = pd.read_csv("../../../data/stage_1/covid_confirmed_usafacts.csv")
covid_confirmed_dataset
```

Defining the variables and datatypes of the covid cases dataset.

```python
covid_trends_df = pd.DataFrame(covid_confirmed_dataset)
covid_trends_df.dtypes
```

Extracting the data from the confirmed covid cases dataset of the California state.

```python
ca_covid_cases = covid_trends_df[covid_trends_df['State'] == 'CA']
ca_covid_cases
```

Extracting the data from the confirmed covid cases dataset of the California state for the last one week.

```python
# selecting the required rows with last week data with state
new_covid_trends_df=covid_trends_df.iloc[: ,[1,2,-7,-6,-5,-4,-3,-2,-1]]
new_covid_trends_df

# choosing state 
nc_covid_cases = new_covid_trends_df[new_covid_trends_df['State'] == 'CA'] 
nc_covid_cases
```

# Observing Trends

Generating a python list containing COVID-19 case sum totals for the individual days in week.

Also printing the total number of cases on each perticular day on the last week (i.e. from 10th Aug 2021 to 16th Aug 2021).

```python
sum_covid_cases = [sum(ca_covid_cases['2021-08-10']), sum(ca_covid_cases['2021-08-11']),
    sum(ca_covid_cases['2021-08-12']), sum(ca_covid_cases['2021-08-13']),
    sum(ca_covid_cases['2021-08-14']), sum(ca_covid_cases['2021-08-15']),
    sum(ca_covid_cases['2021-08-16'])] 

dates = ['2021-08-10','2021-08-11','2021-08-12','2021-08-13','2021-08-14','2021-08-15','2021-08-16']

for y in sum_covid_cases:
    print(y)
```

Generating the plot to get an idea about spread of covid cases in the last week.

The X-axis in the graph shows the dates of the last week and the Y-axis represents the number of cases in California state.

```python
plt.plot(dates, sum_covid_cases)
plt.title('Number of Covid Cases from 2021-08-10 to 2021-08-16 in California \n')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.tick_params(axis='x', labelrotation = 50)
figure = plt.show()
```

According to the graph, the number of cases are increasing in the fisrt four days of the week. Later it got stabilized for the next two days. Finally it got spiked on the last day. Overall the number of cases got increased in the last week.

Reading the hospital bed dataset and displaying the data.

Pre-processing the hospital bed dataset. Dropping rows having NAN in fips_code column. Converted all NAN cells to 0.

```python
frame_display = hospital_beds_dataset_df.drop([0])
frame_display

frame_display_df = pd.DataFrame(frame_display)
frame_display_df

frame_display_df = frame_display_df[frame_display_df['fips_code'].notna()]
frame_display_df = frame_display_df.fillna(0)

frame_display_df
```

Reading and displaying the super Covid-19 dataframe which was exported by merging the cases, deaths and population datasets.

```python
large_covid_data = pd.read_csv('/Users/laharichilakuri/Desktop/DataScience/Project/DS project/data/stage_1/large_covid.csv')
large_covid_data

large_covid_data_df = pd.DataFrame(large_covid_data)
large_covid_data_df
```

# Merge of COVID-19 dataset with Hospital Beds Enrichment dataset

```python
#Merged covid_large_df and hospital_beds_df using inner join 
#by joining on "countyFIPS" in covid_large_df with "fips_code" in hospital_beds_df

COVID_HospitalBeds_merged_df = pd.merge(large_covid_data_df,frame_display_df,how="inner",left_on="countyFIPS",right_on="fips_code")
COVID_HospitalBeds_merged_df.head()
```
