Converting the data from wide format to long format

```python
confirmed_1 = confirmed
confirmed_1 = df = pd.melt(frame = confirmed_1, id_vars=['countyFIPS','County Name','State','StateFIPS'], var_name= ['Date'], value_name= "Cases")
confirmed_1
```

```python

deaths_1 = deaths
deaths_1 = pd.melt(frame = deaths_1, id_vars=['countyFIPS','County Name','State','StateFIPS'], var_name= ['Date'], value_name= "Death")
deaths_1
```

Cleaning the data

```python
population = population.drop('County Name',1)
population.rename(columns = {'population':'Population'}, inplace = True)
```

Merging the data

```python
combined_2 = pd.merge(confirmed_1,deaths_1,
        on = ['countyFIPS','County Name',"State",'StateFIPS',"Date"])
```

```python
final_combined_long= pd.merge(combined_2,population,
        on = ['countyFIPS',"State"])
```

```python
final_combined_long["Date"]= pd.to_datetime(final_combined_long["Date"])
```

```python
final_combined_long
```

```python
final_combined_long.to_csv("../../../data/stage_1/long_large_covid.csv",index = False)
```

```python
 #identify indices with values to remove
ind = population_2[population_2['countyFIPS'] == 0].index

# Drop identified indices from core data
population_2.drop(ind, inplace= True)

# Display clean DataFrame
#population_2

 #identify indices with values to remove
#ind1 = confirmed_2[confirmed_2['countyFIPS'] == 0].index

# Drop identified indices from core data
#confirmed_2.drop(ind1, inplace= True)

# Display clean DataFrame
confirmed_2 = confirmed[confirmed['County Name']!= 'Statewide Unallocated']
confirmed_2
```

```python
#Changin from float to string to help with processing
population_2['countyFIPS'] = deaths['countyFIPS'].astype(str)
population_2['StateFIPS'] = deaths['StateFIPS'].astype(str)

confirmed_2['countyFIPS'] = confirmed['countyFIPS'].astype(str)
confirmed_2['StateFIPS'] = confirmed['StateFIPS'].astype(str)

deaths_2['countyFIPS'] = deaths['countyFIPS'].astype(str)
deaths_2['StateFIPS'] = deaths['StateFIPS'].astype(str)
population_2
```

```python
#get popultation of each county
population_df = population_2[population_2.columns[3:4]]

#get total deaths by county
total_deaths = deaths_2[deaths_2.columns[576:577]]

#get total cases by county 
total_cases = confirmed_2[confirmed_2.columns[576:577]]

#get data that exisits between all dataframes
confirmed_df = confirmed_2[confirmed_2.columns[0:4]]
```

```python
df1 = pd.concat([confirmed_df, total_cases, total_deaths, population_df], axis = 1)

#renaming columns
df1.columns =['countyFIPS', 'County Name', 'State', 'StateFIPS', 'Total Cases', 'Total Deaths','Population']
df1
```

```python
df_final = df1.dropna(how = 'any')
df_final
```

```python
df_final = df_final[df_final['County Name']!='Statewide Unallocated']
df_final
```

```python
df_final.to_csv("../../../data/stage_1/final_covid.csv",index = False)
```
