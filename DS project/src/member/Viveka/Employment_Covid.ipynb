##Reading employment enrichment dataset

```python
employment_data = pd.read_excel("../../../data/stage_1/EmploymentEnrichment1data.xlsx")
employment_dataframe = pd.DataFrame(employment_data)
employment_dataframe

```

```python
employment_dataframe.head()
```

conversion of all the Area names to lower case for consistency

```python

employment_dataframe['Area'] = employment_dataframe['Area'].apply(lambda cn: cn.lower().strip())
```

Selecting only the rows which are of Area type 'County'

```python

employment_dataframe=employment_dataframe.loc[employment_dataframe['Area Type']=='County']
```

```python
employment_dataframe=employment_dataframe.loc[employment_dataframe['Industry'].isin(['10 Total, all industries','1021 Trade, transportation, and utilities','1013 Manufacturing','1024 Professional and business services'])]
```

Removing total covered as we can add up state,federal,local and private count to get that number

```python

employment_dataframe=employment_dataframe.loc[employment_dataframe['Ownership']!='Total Covered']
```

Removing the rows wih status code as 'N' as they are not being updated

```python

employment_dataframe=employment_dataframe.loc[employment_dataframe['Status Code']!='N']
```

```python
employment_dataframe.columns
```

Removing the variables which are not necessary from the dataset

```python

employment_dataframe= employment_dataframe.drop(['St','Cnty','NAICS','Year','Qtr','Area Type','Status Code'], 1)
```

Removing some unknown or undefined data in counties

```python

employment_dataframe = employment_dataframe[employment_dataframe['Area'].str.contains('county', na=False)]
```

```python
employment_dataframe
```

#Displaying the processed dataset

```python

employment_dataframe.head(20)
```

Reading the covid super(large_covid) dataset

```python

covid_superdata = pd.read_csv("../../../data/stage_1/large_covid.csv")
```

Converting object type to int to be able to merge the data with the countyFIPS column which is int

```python

employment_dataframe['Area\nCode']=employment_dataframe['Area\nCode'].astype(str).astype(int)
```

# Merging the covid super dataset and employee enrichment dataset

```python
dataframe = pd.merge(covid_superdata,employment_dataframe,left_on='countyFIPS',right_on='Area\nCode')
```

Removing duplicate columns

```python

dataframe = employment_dataframe.drop(['Area\nCode','Area','St Name'], 1)
```
