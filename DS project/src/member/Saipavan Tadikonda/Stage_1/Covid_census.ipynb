In order to merge the large covid dataset with the census demographic set we need to search for the common variables in both data sets. 

In enrichment dataset there is a column GEO_ID which contains unique five digit countyFIPS code. 

So, slice the GEO_ID to get the countyFIPS as it one of the common variable with the large covid dataset.

Change the datatype of GEO_ID from object to int64 as it should match with datatype in large covid dataset and rename the column name from GEO_ID to countyFIPS

```python
cn_demographic['GEO_ID']= cn_demographic['GEO_ID'].str.slice(start=9)
cn_demographic = cn_demographic.astype({'GEO_ID' : 'int64'})
cn_demographic = cn_demographic[cn_demographic.GEO_ID<72001]
cn_demographic.rename(columns = {'GEO_ID':'countyFIPS'},inplace = True)
cn_demographic
```

Spliting the NAME column in the enrichment dataset which contains both County Name and State

```python
cn_demographic[['County Name','State']]= cn_demographic['NAME'].str.split(',',1,expand=True)
cn_demographic
```

```python
temp1 = pd.read_csv("../../../../data/stage_1/large_covid.csv")
large_covid = pd.DataFrame(temp1)
large_covid
```

Dropping the additional column County Name_y  in the large covid dataset

Renaming the column County Name_x to County Name in order to perform merge.

```python
large_covid.drop(columns=['County Name_y'],axis=1,inplace =True)
large_covid.rename(columns={'County Name_x':'County Name'},inplace =True)
#large_covid = large_covid[large_covid["County Name"]!= "Statewide Unallocated"]
large_covid
```

Finding the datatypes of large covid dataset to confirm whether the common columns are matching with the census demographic dataset

```python
large_covid.dtypes
```

```python
large_covid.shape   #Finding the shape of the large covid dataset
```

Applying strip to County Name columns in the both data sets to remove all the unwanted spaces so that it merges correctly.

```python
cn_demographic.drop(columns=['Total population'],axis=1,inplace=True)
cn_demographic['County Name']= cn_demographic['County Name'].apply(lambda x: x.strip())
large_covid['County Name']= large_covid['County Name'].apply(lambda x: x.strip())
cn_demographic
```

Merging large covid and demographic datasets on countyFIPS, County Name and State(Initial Merge)

```python
final_merge = pd.merge(large_covid,cn_demographic,on=['countyFIPS','County Name','State'])
final_merge
```

Writing the merged data to covid_census csv file 

```python
final_merge.to_csv("../../../../data/stage_1/covid_census.csv",index = False)
```
