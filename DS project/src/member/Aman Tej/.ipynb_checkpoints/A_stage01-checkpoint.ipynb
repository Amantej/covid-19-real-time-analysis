### Importing Pandas library to play around with data.

```python
import pandas as pd
```

Redaing confirmed covid cases dataset.

```python
covid_trend = pd.read_csv("../../../data/stage_1/covid_confirmed_usafacts.csv")
```

Firstly, removing state row where countyFIPS would be '0'. As we are working on county level for analysis which sums up for the state.
- Performing required trend projectio for state - New York.
- Using iloc we get last one week data of all county's present in New York state.

```python
covid_trend = covid_trend[covid_trend["County Name"] != "Statewide Unallocated"]
al_trend = covid_trend[covid_trend["State"] == "NY"].iloc[:,-7:]
al_trend
```

### Plotting the data to observe the trend over last week. 

```python
al_trend.sum().plot(x="Date", title='Covid confirmed cases trend of NY (last week)').set_ylabel("COVID-19 cases")
```

Observations made:
1. Confirmed covid cases in New York tend to increase over time.
2. Over approx. 20k+ cases were increased over this week.
3. Total cases in New York is over 2.19 million.

## Enrichment Data

Reading unchanged/direct downloaded Economic csv enrichment dataset.

```python
en_economic = pd.read_csv("..\..\..\data\stage_1\economic.csv",low_memory= False)
en_economic
```

We have county, state and fip codes in our super covid dataset, which are present in economic enrichment data with other column names. Inorder to perform a merge, we have to modify the enrichment data accordingly and perform the merge to make any analysis further.<br>
<br>
*Spliting the NAME column to both county name and state. Also checking initial datatypes of the data.*

```python
en_economic[['County Name','State']] = en_economic['NAME'].str.split(',',1,expand=True)
en_economic.dtypes
```

Economic data set is huge, so we can split by specific columns to perform a particular level of analysis and can remove unwanted columns which would not be considered for analysis. This is an inital merge, later I may remove/add few columns accordingly..!

```python
en_economic = en_economic[['GEO_ID','County Name','State','DP03_0018E','DP03_0019E','DP03_0020E','DP03_0021E','DP03_0022E','DP03_0023E','DP03_0024E','DP03_0025E','DP03_0026E','DP03_0027E','DP03_0028E','DP03_0029E','DP03_0030E','DP03_0031E','DP03_0097E','DP03_0098E','DP03_0099E','DP03_0051E', 'DP03_0063E','DP03_0088E']]
en_economic.head()
```

*Renaming the columns for better understanding of particular column and dropping description row*

```python
en_economic = en_economic.rename(columns={'County Name':'County Name','State':'State','DP03_0018E':'Total people commuting to work','DP03_0019E':'Driving alone','DP03_0020E':'Carpooled','DP03_0021E':'Only public transports','DP03_0022E':'Walked','DP03_0023E':'Other means','DP03_0024E':'Work from home','DP03_0025E':'Mean travel time to work',
'DP03_0026E':'Total people has occupation','DP03_0027E':'Management&business&science&arts occupations','DP03_0028E':'Service occupations','DP03_0030E':'Natural resources, construction, and maintenance occupations','DP03_0029E':'Sales and office occupations','DP03_0031E':'Production, transportation, and material moving occupations',
'DP03_0097E':'Total people has health insurances','DP03_0097E':'Private health insurance','DP03_0098E':'Public health insurance','DP03_0099E':'No health insurance','DP03_0051E':'Total households', 'DP03_0063E':'Mean household income','DP03_0088E':'Per capita income'})
```

```python
en_economic.drop(en_economic.index[0], inplace = True)
en_economic
```

Geo Id is countyFIP code, so making the long id to 5 digit FIP code using slicing.

```python
en_economic['GEO_ID'] = en_economic['GEO_ID'].str.slice(start=9)
en_economic.rename(columns={"GEO_ID":"countyFIPS"},inplace=True)
en_economic = en_economic.astype({'countyFIPS': 'int64'})
```

```python
state_map = {' Alaska' : 'AK',
' Alabama' : 'AL',
' Arkansas' : 'AR',
' American Samoa' : 'AS',
' Arizona' : 'AZ',
' California' : 'CA',
' Colorado' : 'CO',
' Connecticut' : 'CT',
' District of Columbia' : 'DC',
' Delaware' : 'DE',
' Florida' : 'FL',
' Georgia' : 'GA',
' Guam' : 'GU',
' Hawaii' : 'HI',
' Iowa' : 'IA',
' Idaho' : 'ID',
' Illinois' : 'IL',
' Indiana' : 'IN',
' Kansas' : 'KS',
' Kentucky' : 'KY',
' Louisiana' : 'LA',
' Massachusetts' : 'MA',
' Maryland' : 'MD',
' Maine' : 'ME',
' Michigan' : 'MI',
' Minnesota' : 'MN',
' Missouri' : 'MO',
' Northern Mariana Islands' : 'MP',
' Mississippi' : 'MS',
' Montana' : 'MT',
' National' : 'NA',
' North Carolina' : 'NC',
' North Dakota' : 'ND',
' Nebraska' : 'NE',
' New Hampshire' : 'NH',
' New Jersey' : 'NJ',
' New Mexico' : 'NM',
' Nevada' : 'NV',
' New York' : 'NY',
' Ohio' : 'OH',
' Oklahoma' : 'OK',
' Oregon' : 'OR',
' Pennsylvania' : 'PA',
' Puerto Rico' : 'PR',
' Rhode Island' : 'RI',
' South Carolina' : 'SC',
' South Dakota' : 'SD',
' Tennessee' : 'TN',
' Texas' : 'TX',
' Utah' : 'UT',
' Virginia' : 'VA',
' Virgin Islands' : 'VI',
' Vermont' : 'VT',
' Washington' : 'WA',
' Wisconsin' : 'WI',
' West Virginia' : 'WV',
' Wyoming' : 'WY',
}
```

Mapping state names to two letter codes as it matches with super covid dataset. <br>
(Note: If you find NaN please run the code from very beginning.)

```python
en_economic['State'] = en_economic.State.map(state_map) 
en_economic.dtypes
```

Reading large_covid dataset.

```python
super_covid = pd.read_csv("../../../data/stage_1/large_covid.csv")
super_covid.drop(columns = ['County Name_y'],axis=1,inplace=True)
super_covid.rename(columns = {'County Name_x':'County Name'},inplace=True)
super_covid.head()
```

Performing strip to extra spaces in county name column in both datasets to merge correctly.

```python
en_economic['County Name'] = en_economic['County Name'].apply(lambda cn: cn.strip())
super_covid['County Name'] = super_covid['County Name'].apply(lambda cn: cn.strip())
```

Inner merge is performed on super covid dataset and economic enrichment dataset on basis of countyname, state, fip codes.

```python
inital_merge = pd.merge(super_covid, en_economic, on = ['County Name','State','countyFIPS'])
```

```python
inital_merge.head(5)
```

Writing the merged data to economic_covid.csv file for further usage.

```python
inital_merge.to_csv("../../../data/stage_1/economic_covid.csv")
```
