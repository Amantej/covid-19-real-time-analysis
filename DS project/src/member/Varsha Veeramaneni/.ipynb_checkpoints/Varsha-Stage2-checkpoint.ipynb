## Generating weekly statistics (mean, median, mode) for number of new cases and deaths across the Virginia state

Reading the Covid dataset

```python
data = pd.read_csv("../../../data/stage_1/long_large_covid.csv")
data
```

Removing the unallocated states

```python
data= data[data['County Name'] != 'Statewide Unallocated']
data
```

converting all dates from object formats to datetime

```python
data['Date'] = data.Date.apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
```

```python
# choosing the state
VA_state = data[data['State'] == "VA"]
VA_state.head()
```

Finding the new cases and deaths per day

```python
VA_state['New Cases Per Day'] = VA_state['Cases'].diff()
VA_state['New Deaths Per Day'] = VA_state['Death'].diff()
VA_state
```

Droping the first date as we we are using .diff, we will get the first value as NAN 

```python
VA_state =VA_state[VA_state.Date != '2020-01-22']
```

```python
VA_state
```

```python
# group by Date
VA_data = VA_state.groupby(["Date"]).sum().reset_index()
VA_data
```

```python
#Mean of the cases and deaths
VA_data_mean = VA_data.groupby([ pd.Grouper(key='Date', freq='w')])['New Cases Per Day','New Deaths Per Day'].agg({'New Cases Per Day':'mean',
                                                                                                'New Deaths Per Day' :'mean'}).round().reset_index()
VA_data_mean
```

finding the mean,median and mode for number of cases for Virginia State

```python
mean = VA_data_mean["New Cases Per Day"].agg(["mean"])
median = VA_data_mean["New Cases Per Day"].agg(["median"])
mode = VA_data_mean["New Cases Per Day"].mode()

print('Virginia: ', '\n', 'Weekly Case Mean: ', mean[0], '\n', 'Weekly Case Median: ', median[0], '\n', 'Weekly Case Mode: ', mode[0])
      
```

finding the mean,median and mode for number of deaths for Virginia State

```python
mean = VA_data_mean["New Deaths Per Day"].agg(["mean"])
median = VA_data_mean["New Deaths Per Day"].agg(["median"])
mode = VA_data_mean["New Deaths Per Day"].mode()
print('Virginia: ', '\n', 'Weekly Death Mean: ', mean[0], '\n', 'Weekly Death Median: ', median[0], '\n', 'Weekly Death Mode: ', mode[0])
```

```python
#Normalizing the population 
VA_data['New Cases Per Day Normalized']= (VA_data['New Cases Per Day']/VA_data['Population'])*100000
VA_data['New Deaths Per Day Normalized']= (VA_data['New Deaths Per Day']/VA_data['Population'])*100000
VA_data
```

```python
#Mean of the Normalized cases and deaths
VA_data_1 = VA_data.groupby([ pd.Grouper(key='Date', freq='w')])['New Cases Per Day Normalized','New Deaths Per Day Normalized'].agg({'New Cases Per Day Normalized':'mean',
                                                                                                'New Deaths Per Day Normalized' :'mean'}).round().reset_index()
VA_data_1
```

finding the mean,median and mode for cases which are normalized by population  

```python
mean = VA_data_1["New Cases Per Day Normalized"].agg(["mean"])
median = VA_data_1["New Cases Per Day Normalized"].agg(["median"])
mode = VA_data_1["New Cases Per Day Normalized"].mode()

print('Virginia: ', '\n', 'Weekly Case Mean: ', mean[0], '\n', 'Weekly Case Median: ', median[0], '\n', 'Weekly Case Mode: ', mode[0])
      
```

finding the mean,median and mode for deaths which are normalized by population  

```python
mean = VA_data_1["New Deaths Per Day Normalized"].agg(["mean"])
median = VA_data_1["New Deaths Per Day Normalized"].agg(["median"])
mode = VA_data_1["New Deaths Per Day Normalized"].mode()
print('Virginia: ', '\n', 'Weekly Death Mean: ', mean[0], '\n', 'Weekly Death Median: ', median[0], '\n', 'Weekly Death Mode: ', mode[0])
```

## Compare the data against other states

Choosing five other states along with Virginia

```python
states= data[data["State"].isin(["NC", "FL", "NY", "TX", "CA","VA"])]
states
```

```python
states['New Cases Per Day'] = states['Cases'].diff()
states['New Deaths Per Day'] = states['Death'].diff()
states
```

```python
num = states['New Cases Per Day']._get_numeric_data()
num[num < 0] = 0
num1 = states['New Deaths Per Day']._get_numeric_data()
num1[num1 < 0] = 0
```

```python
#Droping the first date as we we are using .diff, we will get the first value as NAN 
states= states[states['Date'] != '2020-01-22']
```

## Calculate total number of normalized number of cases and deaths in each state for all the days



```python
# Group by State and Date
states = states.groupby(["State","Date"]).sum().reset_index()
states
```

Normalize the cases and deaths  by population(per 100000)

```python
states['New Cases Per Day Normalized']= (states['New Cases Per Day']/states['Population'])*100000
states['New Deaths Per Day Normalized']= (states['New Deaths Per Day']/states['Population'])*100000
states
```

Calculate the weekly number of cases and deaths for each state

```python
#Grouping by week and finding the mean
states_weekly = states.groupby(["State", pd.Grouper(key='Date', freq='w')])['New Cases Per Day Normalized','New Deaths Per Day Normalized'].agg({'New Cases Per Day Normalized':'mean',
                                                                                                'New Deaths Per Day Normalized' :'mean'}).reset_index()
states_weekly
```

Calculate the weekly mean and median of the number of cases

```python
states_weekly.groupby('State')['New Cases Per Day Normalized'].agg(['mean','median']).reset_index()
```

```python
#Getting mode of cases of all 5 states
states_weekly.groupby('State')['New Cases Per Day Normalized'].agg(lambda x:x.value_counts().index[0]).reset_index()
```

## Comparision
From the above figure I obseve that the Weekly mean number of cases for Florida is the highest and the weekly mean number of cases for Virginia is the lowest among the 6 states.

Weekly Median number of cases is the highest for Florida and lowest for California

```python

fig1_VA = px.line(states_weekly, x="Date", y="New Cases Per Day Normalized", color='State', title="Weekly cases trend of 6 states")
fig1_VA.write_image("../../../Images/member/Varsha/stage_2/fig1_VA.png")
img_bytes = fig1_VA.to_image(format="png")
Image(img_bytes)
```

From the above figure, I observed that though Texas state has less Weekly mean number of cases than Florida, at some point, the rise in cases is very high than in Florida.  And remaining states follow the same pattern though the numbers are different.

Calculate the weekly mean,median and mode of the number of deaths

```python
states_weekly.groupby('State')['New Deaths Per Day Normalized'].agg(['mean','median']).reset_index()
```

```python
#Getting mode of deaths of all states
states_weekly.groupby('State')['New Deaths Per Day Normalized'].agg(lambda x:x.value_counts().index[0]).reset_index()
```

## Comparision

From the above figure I obseve that the Weekly mean number of deaths for New York is the highest and the weekly mean number of deaths for NC is the lowest among the 6 states.

Weekly Median number of deaths is the highest for Florida and lowest for New York

```python

fig2_VA = px.line(states_weekly, x="Date", y="New Deaths Per Day Normalized", color='State', title="Weekly deaths trend of 6 states")
fig2_VA.write_image("../../../Images/member/Varsha/stage_2/fig2_VA.png")
img_bytes = fig2_VA.to_image(format="png")
Image(img_bytes)
```

In the plot of deaths among the six states, I have observed that New York state has highest number of death rates followed by Texas when we compared to the other states. While the rest four states follow the similar pattern among which almost have their highest peak of deaths in between January and April. 

## Do they compare with the US pattern?
When we compare the above plots with the US plots, it can be said that the cases follow a similar pattern, where both have the highest peak point in January, and after that, there is a fall in the graph, and again from June, the cases increased slightly. 
In the plots of the deaths, both figures follow the almost same pattern, where the number of deaths is high in May and later in January of the following year.


## Identify counties within the previous state with high case and death rates. (normalize by population)

```python
VA_state
```

```python
#Groupby counties and calculate the sum(total) number of cases per day and number of deaths cases per day
VA_data_county=VA_state.groupby('County Name').agg({'countyFIPS':'first','Population':'first','New Cases Per Day':'sum','New Deaths Per Day':'sum'}).reset_index()
VA_data_county
```

```python
VA_data_county = VA_data_county[VA_data_county['County Name'] != "statewide unallocated"]
```

normalize the total number of cases and deaths of each county by dividing it by total population and multiply by 10000

```python
VA_data_county['cases'] = (VA_data_county['New Cases Per Day']/VA_data_county['Population'])*10000
VA_data_county['deaths'] = (VA_data_county['New Deaths Per Day']/VA_data_county['Population'])*10000
VA_data_county
```

```python
#top 5 counties with high case rates
VA_cases_high=VA_data_county.sort_values(['cases'],ascending=False).head()
VA_cases_high.head()
```

```python
#top 5 counties with high death rates
VA_deaths_high=VA_data_county.sort_values(['deaths'],ascending=False).head()
VA_deaths_high.head()
```
