Reading the long large covid csv file that is generated in the stage_1(team task)

```python
cases_df = pd.read_csv('../../../../data/stage_1/long_large_covid.csv')
cases_df.head()
```

Selecting the data related to the state NC and moving into separate dataframe

```python
cases_df_nc = cases_df[cases_df['State']== 'NC']
cases_df_nc
```

Removing rows whose countyFIPS is equal to 0

```python
cases_df_nc= cases_df_nc[cases_df_nc['countyFIPS']!= 0]
cases_df_nc
```

Finding the new cases and new deaths every day and removing the data related to the date 2020-01-22

```python
cases_df_nc['new_cases'] = cases_df_nc['Cases'].diff()
cases_df_nc['new_deaths']= cases_df_nc['Death'].diff()
cases_df_nc = cases_df_nc[cases_df_nc.Date!= '2020-01-22']
cases_df_nc
```

Grouping the data by date using groupby and calculating sum for new_cases and new_deaths for that particular date

```python
cases_df_nc_1 = cases_df_nc[['Date','new_cases','new_deaths']]
cases_df_nc_1 = cases_df_nc_1.groupby('Date').sum().reset_index()
cases_df_nc_1_cases = cases_df_nc_1[['Date','new_cases']]
cases_df_nc_1_cases
```

In order to get the start day of the infection, I took the new_cases into a list and wrote a for loop to find out the index of the list where the value is greater than or equal to 1.

Then I wrote another for loop to append the count from the index which I found above and appended the each count to the each row from the start day of the infection

```python
list_new_cases = list(cases_df_nc_1_cases['new_cases'])  #into list
c1 = []
c2 = []
count = 1
for i in list_new_cases:
    if int(i) >= 1:
        c1= list_new_cases.index(i)       #finding index
        break
        
for i in range(len(list_new_cases)):
    if(i >= c1 and i!= 'NaN'):            #greater than or equal to index
        c2.append(count)                  #appending count
        count = count+1                   #incrementing the count by 1
    else:
        c2.append(0)
```

Reference for dropping particular rows using iloc https://stackoverflow.com/questions/60698147/how-to-drop-a-row-using-iloc-method

Creating a new column start_day that stores the count values which we found above. And drop the records whose start_day is zero as they are not the start date of the infection

```python
cases_df_nc_1_cases['start_day'] = c2
cases_df_nc_1_cases.drop(index = cases_df_nc_1_cases.iloc[:c1,:].index.tolist(), inplace=True)
cases_df_nc_1_cases
```

Applying the linear regression model where X is the count from the start day of infection and y is new cases on a particular day

```python
X = cases_df_nc_1_cases[['start_day']]    #passing the start day column
y = cases_df_nc_1_cases['new_cases']      #new_cases on a particular day
lm = LinearRegression()
regression = lm.fit(X, y)                 #fitting a linear regression model
preds = regression.predict(X)             #predictions on the trained values
X1 = np.arange(len(X)+1,len(X)+8).reshape(-1, 1)      # getting the next 7 days 
predict1 = regression.predict(X1)                   #prediction for the added next 7 days
plt.figure(figsize=(10,8))
plt.title("Linear regression for new_cases")
plt.scatter(X, y, color='red')
plt.plot(X, preds , color='blue', linewidth = 3, label='Linear Regression')
plt.plot(X1, predict1 , color='blue', linewidth = 7, linestyle = 'dotted', label='Linear Regression for next 7days')
print('Next 7 days predictions for Linear Regression:', predict1)

plt.legend()
```

Calculating the root mean square error for Linear regression

```python
rmse = mean_squared_error(y,preds, squared = False)
print('root mean square error for Linear Regression is:',rmse)
```

i)	Alternative Hypothesis: Covid cases are higher where there are a greater number of males?

Null Hypothesis: Covid cases are normal where there are greater number of males?
    
After applying the two tail two sample t-test the obtained t-statistic value and pvalue are as follows:

```python
stats.ttest_ind(a= merged['Males'],
                b= merged['Total Cases'] )
```

From the above result we can see that p-value is less than the threshold value i.e., 0.05 so we can reject the null hypothesis.

Now we need to apply one-tail-two sample t-test to check whether it lies on lower side or greater side.

From the above result we can see that p/2 < 0.05 and t-statistic value >0. So, it lies on the greater side.

Null Hypothesis: Covid cases are less where there are a greater number of white population?

```python
stats.ttest_ind(a= merged['White population'],
                b= merged['Total Cases'])
```

From the above result we can see that p-value is less than the threshold value i.e., 0.05 so we can reject the null hypothesis.

Now apply one-tail-two sample t-test to check whether it lies on lower side or greater side.
