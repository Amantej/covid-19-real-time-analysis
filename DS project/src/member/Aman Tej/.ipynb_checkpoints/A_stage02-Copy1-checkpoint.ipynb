### Task 1.a Generate weekly statistics (mean, median, mode) for number of new cases and deaths across a specific state.

 #### Cases Statistics for NY

Reading Covid cases dataset to filter specific state

```python
covid_cases = pd.read_csv('../../../data/stage_2/covid_confirmed_usafacts.csv')
covid_cases.head()
```

Removing state rows and filtered data for state 'Newyork'.

```python
covid_cases = covid_cases[covid_cases['countyFIPS'] != 0]
covid_cases = covid_cases[covid_cases['State'] == 'NY']
covid_cases.head()
```

Removing first four columns as they do not contribute to our requirement. And taking sum, row wise to understand total cummilative cases count on a given date.

```python
filter_data = covid_cases.iloc[:, 4:].sum()
filter_data_1 = pd.DataFrame(columns=covid_cases.columns[4:])
filter_data_1 = filter_data_1.append(filter_data, ignore_index=True)
filter_data_1
```

To get individual day cases, we are using diff function and making column wise difference and removing first date as it gives NaN

```python
filter_data_1 = filter_data_1.diff(axis=1)
filter_data_1.drop(columns=['2020-01-22'], inplace=True)
filter_data_1
```

Taking rounded total weeks value generated by given cases data and coverting given date, cases into weekly cases count. Also, appended column names as week starting date. 

```python
l = []
for i in range(len(filter_data_1.columns)//7+1): l.append(filter_data_1.iloc[:,i*7:i*7+7].sum(axis=1).astype(int)) 
```

```python
cases_weekly = pd.concat(l, axis=1)
cases_weekly = cases_weekly.set_axis(filter_data_1.columns[:(len(filter_data_1.columns)//7+1)*7:7],axis=1)
cases_weekly
```

Cal. Mean for individual weeks (sum of cases) 

```python
round(cases_weekly.mean(axis=1)).astype(int)
```

Cal. Median for individual weeks (sum of cases) 

```python
round(cases_weekly.median(axis=1)).astype(int)
```

Cal. Mode for individual weeks (sum of cases) 

```python
round(cases_weekly.mode(axis=1)).astype(int)
```

Taking rounded total weeks value generated by given cases data and coverting given date, cases into weekly cases mean. Also, appended column names as week starting date. 

```python
k = []
for i in range(len(filter_data_1.columns)//7+1): k.append(round(filter_data_1.iloc[:,i*7:i*7+7].mean(axis=1)).astype(int))
```

```python
cases_weekly_mean = pd.concat(k, axis=1)
cases_weekly_mean = cases_weekly_mean.set_axis(filter_data_1.columns[:(len(filter_data_1.columns)//7+1)*7:7],axis=1)
cases_weekly_mean
```

Cal. Mean for individual weeks (mean of cases) 

```python
round(cases_weekly_mean.mean(axis=1)).astype(int)
```

Cal. Median for individual weeks (mean of cases) 

```python
round(cases_weekly_mean.median(axis=1)).astype(int)
```

Cal. Mode for individual weeks (mean of cases) 

```python
round(cases_weekly_mean.mode(axis=1)).astype(int)
```

#### Filtering New York state data from stage_1 merged data

Reading large format covid merged data from stage_1 to reduce melt operations.

```python
states_data = pd.read_csv('../../../data/stage_1/long_large_covid.csv')
states_data.head()
```

Filtering on specific states, which are closely populated as Newyork.

```python
states_data = states_data[states_data["State"].isin(['NY','CA','TX','FL','IL','OH'])].reset_index(drop = True)
```

Removing state rows

```python
states_data = states_data[states_data['countyFIPS'] != 0]
states_data
```

Applying datetime strptime to date conversion.

```python
from datetime import datetime
states_data['Date'] = states_data.Date.apply(lambda day: datetime.strptime(day, "%Y-%m-%d"))
```

To get exact new cases count for a particular date we have to use diff. function and remove starting date for every state.

```python
states_data['new_cases'] = states_data['Cases'].diff()
states_data['new_deaths'] = states_data['Death'].diff()
states_data= states_data[states_data['Date'] != '2020-01-22']
states_data
```

To get total count of cases on a particular date in state. We are grouping by state and date.

```python
states_data1 = states_data.groupby(["State","Date"]).sum().reset_index()
states_data1
```

Here all the states we took before has 15 million to 40 million population (approx.), so we are taking normalization per 100000 according to population.

```python
states_data1['Normalized cases'] = (states_data1['new_cases']/states_data1['Population'])*100000
states_data1['Normalized deaths'] = (states_data1['new_deaths']/states_data1['Population'])*100000
states_data1
```

Strting with monday 2021-01-27 to get full week formations and ending on sunday i.e., 2021-08-15. 

```python
states_data1 = states_data1[(states_data1['Date'] >= '2020-01-27') & (states_data1['Date'] <= '2021-08-15')]
```

Splitting by weeks for every state and summing all the cases in that week.

```python
states_data_weekly = states_data1.groupby('State').resample('W', on='Date').mean().reset_index().sort_values(by=['State','Date'])
states_data_weekly
```

Mean and median for every states total normalized cases per 100000.

```python
states_data_weekly_sum_mean = states_data_weekly.groupby('State')['Normalized cases'].agg(['mean','median']).reset_index().round()
states_data_weekly_sum_mean['Population'] = states_data1['Population'].unique()
states_data_weekly_sum_mean
```

importing plotly to plot line plots.

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as pgo
from IPython.display import Image
#!pip install -U kaleido
```

line plot for every states weekly normalized cases per 100000.

```python
normalized_cases_plot = px.line(states_data_weekly, x ='Date', y='Normalized cases', color='State')
normalized_cases_plot.write_image('../../../Images/member/Amantej/stage_2/normalized_cases_plot.png')
f1 = normalized_cases_plot.to_image(format="png")
Image(f1)
```

Mean and median for every states total normalized deaths per 100000.

```python
states_data_deaths_weekly_sum_mean = states_data_weekly.groupby('State')['Normalized deaths'].agg(['mean','median']).reset_index()
#states_data_deaths_weekly_sum_mean = states_data_weekly.groupby('State')['Normalized deaths'].agg(['mean','median']).reset_index().round()
states_data_deaths_weekly_sum_mean['Population'] = states_data1['Population'].unique()
states_data_deaths_weekly_sum_mean['rounded_mean'] = round(states_data_deaths_weekly_sum_mean['mean'])
states_data_deaths_weekly_sum_mean['rounded_median'] = round(states_data_deaths_weekly_sum_mean['median'])
states_data_deaths_weekly_sum_mean
```

line plot for every states weekly normalized deaths per 100000.

```python
normalized_deaths_plot = px.line(states_data_weekly, x ='Date', y='Normalized deaths', color='State')
normalized_deaths_plot.write_image('../../../Images/member/Amantej/stage_2/normalized_deaths_plot.png')
f2 = normalized_deaths_plot.to_image(format="png")
Image(f2)
```
