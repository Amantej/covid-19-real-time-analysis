## 1. Calculate COVID-19 data trends for last week of the data. Are the cases increasing, decreasing, or stable? Each student chooses a state to analyze.

Import pandas library to read and perform analysis on the csv files.

```python
import pandas as pd
```

Read the covid confirmed csv file and covid deaths csv file and display the data.

```python
temp = pd.read_csv("../../../../data/stage_1/covid_confirmed_usafacts.csv")
confirmed_cases = pd.DataFrame(temp)
temp1 = pd.read_csv("../../../../data/stage_1/covid_deaths_usafacts.csv")
deaths = pd.DataFrame(temp1)
confirmed_cases
```

```python
deaths
```

Excluding the records which has county name as Statewide Unallocated as its countyFIPS is 0.

```python
confirmed_cases_temp = confirmed_cases[confirmed_cases["County Name"]!= "Statewide Unallocated"]
confirmed_cases_temp
```

iloc() function enables us to get the particular cells from the dataset.

Selected NC state for observing the covid confirmed cases trend

Using the iloc function inorder to get the last week data of all counties in the NC state

```python
confirmed_cases_nc = confirmed_cases_temp[confirmed_cases_temp["State"]=="NC"]
confirmed_cases_nc_lastweek = confirmed_cases_nc.iloc[:,-7:]
confirmed_cases_nc_lastweek
```

Adding the cases across all the counties in NC state by date

```python
confirmed_cases_nc_lastweek_sum = confirmed_cases_nc_lastweek.sum(axis=0)
confirmed_cases_nc_lastweek_sum
```

Plotting the data to observe the last week trend of covid confirmed cases by using plot() function.

```python
confirmed_cases_nc_lastweek_sum.plot().set_ylabel("Number of cases")
```

From the above plot we can observe that in the last week the covid cases increased initially and remained stable for three days and again on the last day cases got increased.

Approximately 33,000 cases are registered in the last week.

As of 2021-08-16 the confirmed covid cases in NC are over 1.12 million.
