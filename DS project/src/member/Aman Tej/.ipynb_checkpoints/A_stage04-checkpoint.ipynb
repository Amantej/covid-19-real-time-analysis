### Utilize Linear and Non-Linear (polynomial) regression models to compare trends for a single state and its counties (top 5 with highest number of cases). Start your data from the first day of infections.

Importing all libraries that are used for these tasks.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from scipy.stats import norm
from scipy import stats 
from sklearn.metrics import mean_squared_error 
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

This function takes the state/county dataset and cleans the data. Then returns new cases starting from first day along with date.

```python
def data_clean(df, x, county = None, state = False ):
    temp = df
    temp = temp[temp['countyFIPS'] != 0]
    if county:
        temp = temp[temp['countyFIPS'] == county]
        #print(temp)
    else:
        temp = temp[temp['State'] == state]
        #print(state)
    temp = temp.drop(columns = temp.columns[:4])
    temp_sum = temp.sum()
    temp.drop(temp.index[:], inplace=True)
    new_temp = temp.append(temp_sum, ignore_index=True)
    temp1 = pd.melt(new_temp, var_name = ['Date'], value_name = x)
    temp1[x] = temp1[x].diff()
    temp1[x] = abs(temp1[x])
    temp1 = temp1.fillna(0)
    day_list = list(temp1[x])
    day_count = []
    day = 1
    s = False
    for i in day_list:
        if int(i) >= 1 or s == True:
            s = True
            day_count.append(day)
            day = day+1
        else:
            day_count.append(0)
    temp1['on_particula_day'] = day_count       
    y = day_count.index(1)
    temp1.drop(temp1.iloc[:y,:].index.tolist(), inplace=True)
    return temp1
```

we have pass state name along with required column name as var(cases/deaths)

Confidence interval concept Reference: https://medium.com/swlh/ds001-linear-regression-and-confidence-interval-a-hands-on-tutorial-760658632d99, lecture notebooks.

This is hybrid function, can be used for multiple inputs:
- if we call reg(df,var), only return Linear regression models plots
- if we call reg(df,var,reg_type), returns Linear and Non-linear regression models plots
- if we call reg(df,var,reg_type,pred), returns Linear and Non-linear regression models plots along with default 7 days predictions
Note: Using matplotlib because can simply add multiple traces and which has more documentation. Also, specifically never mentioned about using a specific library.

```python
def reg(df, var, reg_type = False, pred = False, days = 7):
    X = np.stack(df['on_particula_day']).reshape(-1,1)
    y = np.stack(df[var]).reshape(-1,1)
    reg = LinearRegression().fit(X,y)
    predict = reg.predict(X)
    plt.figure(figsize=(20,8))
    plt.title("Linear regression")
    plt.subplot(1, 2, 1)
    plt.scatter(X, y, color='blue',alpha=0.2)
    plt.plot(X, predict , color='black', linewidth = 2, label='Linear Regression')
    plt.legend()
    if(pred):
        X1 = np.arange(len(X),len(X)+days).reshape(-1, 1)
        new_predict = reg.predict(X1)
        print('One week predictions Linear Regression:', list(new_predict),'\n')
        plt.subplot(1, 2, 2)
        plt.scatter(X, y, color='blue',alpha=0.2)
        plt.plot(X, predict , color='black', linewidth = 2, label='Linear Regression', alpha= 0.5)
        plt.plot(X1, new_predict , color='purple',linestyle='dotted', linewidth= 2, label='Predicted with linear reg',alpha=5)
        plt.legend()
    print('RMSE for Linear Regression:', np.sqrt(mean_squared_error(y,predict)),'\n')
    c = {2:'red',3:'green',4:'orange',5:'purple'}
    if(reg_type):
        fig = plt.figure('new')
        plt.figure(figsize=(20,8))
        plt.title("Non-Linear (polynomial) regression")
        plt.subplot(2, 2, 1)
        plt.scatter(X, y, color='blue',alpha=0.2)
        for i in range(2,6): #1 is linear reg. and taking degree till 4, 5 may overfit.
            preg = PolynomialFeatures(degree=i)
            predict1 = preg.fit_transform(X)
            predict_total = LinearRegression().fit(predict1,y)
            plt.plot(X, predict_total.predict(predict1) , color= c[i], linewidth = 2, label='Poly Linear Regression with degree %d'%i, alpha=2)
            print('RMSE for Polynomial Regression with degree %d:'%i, np.sqrt(mean_squared_error(y,predict_total.predict(predict1))),'\n')
        plt.legend()
        if pred:
            plt.subplot(2, 2, 2)
            plt.scatter(X, y, color='blue',alpha=0.2)
            
            for i in range(2,6):
                preg = PolynomialFeatures(degree=i)
                predict1 = preg.fit_transform(X)
                predict_total = LinearRegression().fit(predict1,y)
                predict2 = preg.fit_transform(X1)
                print('One week predictions for Poly Linear Regression with degree %d'%i, list(predict_total.predict(predict2)),'\n')
                plt.plot(X, predict_total.predict(predict1) , color= c[i], linewidth = 2, label='Poly Linear Regression with degree %d'%i, alpha=0.5)
                plt.plot(X1, predict_total.predict(predict2) , linestyle='dotted', color= c[i], linewidth = 2, label='Predicted with model degree %d'%i)    
        plt.legend()
    return reg.coef_
```

This function will give a confidence interval using standard dev and a guassian cruve.

```python
def con_inv(df, var, deg):
#df = county_cases[0]
    X = np.stack(df['on_particula_day']).reshape(-1,1)
    y = np.stack(df[var])
    if(deg == 1):
        reg = LinearRegression().fit(X,y)
        predict = reg.predict(X)
    else:
        preg = PolynomialFeatures(degree=deg)
        predict1 = preg.fit_transform(X)
        predict_total = LinearRegression().fit(predict1,y)
        predict = predict_total.predict(predict1)
    def get_prediction_interval(prediction, y, predict, pi=.95):
        sum_errors = np.sum((y - predict)**2)
        stdev = np.sqrt(1 / (len(y) - 2) * sum_errors)
        op = 1 - pi
        p = 1 - (op / 2)
        z_score = norm.ppf(p)
        interval = z_score * stdev
        lower, upper = prediction - interval, prediction + interval
        return lower, prediction, upper
    lower_val = []
    upper_val = []
    for i in predict:
        lower, prediction, upper =  get_prediction_interval(i, y, predict)
        lower_val.append(lower)
        upper_val.append(upper)
    plt.figure(figsize=(20,8))
    plt.fill_between(np.arange(0,len(y),1),upper_val, lower_val, color='r',label='Confidence Interval',alpha = 0.3)
    plt.scatter(X,y,color='b',alpha = 0.5)
    plt.plot(predict,'k',label='Linear regression')
    plt.title('95% confidence interval')
    plt.legend()
    plt.show()
```

#### Utilize the hospital data to calculate the point of no return for a state. Use percentage occupancy / utilization to see which states are close and what their trend looks like.

Taking all the states filtered data. Which is cases/deaths count on a particular day starting from day 1 and storing the dataframe in state_fil list.

```python
states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
state_fil = []
for i in states:
    state_fil.append(data_clean(df = pd.read_csv('../../../data/stage_1/covid_deaths_usafacts.csv'), x = 'Deaths', county = None, state = i ))
    
```

Now applying linear and non-linear models to predict 15 days from the last day for all states filtered data. And, saving the best model maybe linear or non-linear with degree 2/3/4, this is on the basis of RMSE values of each model. Now, we save the predictions of best model in a list

```python
best = []
for i in state_fil:
    X = np.stack(i['on_particula_day']).reshape(-1,1)
    y = np.stack(i['Deaths']).reshape(-1,1)
    #print('RMSE for Linear Regression:', np.sqrt(mean_squared_error(y,predict)),'\n')
    val = 0
    temp = []
    for i in range(1,5):
        preg = PolynomialFeatures(degree=i) #adding polynomial features to X
        predict1 = preg.fit_transform(X) 
        #print(predict1)
        predict_total = LinearRegression().fit(predict1,y) #fitting model
        X1 = np.arange(len(X),len(X)+15).reshape(-1, 1) #for 10 days predictions
        predict2 = preg.fit_transform(X1)
        predict = predict_total.predict(predict2) #making predictiond
        temp.append(predict)
        if(val > np.sqrt(mean_squared_error(y,predict_total.predict(predict1)))):
            val = i
    #print(temp)
    best.append(temp[val])
state_dict = {}
for i in range(len(best)):
    state_dict.update({states[i]: best[i]})
dict(itertools.islice(state_dict.items(), 5)) #diplaying first 5 dictionary content
```

#### Perform hypothesis tests on questions identified in Stage III. 
e.x. Does higher employment data (overall employment numbers) lead to higher covid case numbers or more rapid increase in covid cases.. Here you would compare the covid cases to the state or county level enrichment data to prove or disprove your null hypothesis. In this case there will be a two tail - two sample t-test to see if there is a difference and then one-tail - two sample t-test to show higher or lower.
#### Depending on your type of data you can also perform Chi-square test for categorical hypothesis testing.

In stage 03, while getting correlation. I used state - NC and saved that as a dataset and using same dataset to prove hypothesis. Beacuse NC has good correlation between cases and insurance columns compared to other states.

```python
enr_data =  pd.read_csv('../../../data/stage_1/NC-enrich.csv')
enr_data
```

1.
<b>Null Hypothesis H0</b>: Not health insurance does not influence increase in covid cases. <br>
<b>Alternative Hypothesis </b> H1: Not having health insurance influence increase in covid cases.

```python
var1 = enr_data['No health insurance']
var2 = enr_data['new_cases']
stats.ttest_ind(var2, var1, alternative='two-sided')
```

We does not reject the null hypothesis and do not accept the alternative hypothesis since the p-value is less than 0.05 (95 percent significance levelc). p/2 = 0.45 > 0.05 which implies our actual hypothesis is false. 

2.
<b>Null Hypothesis H0</b>: Having private health insurance does not influence decreases in covid cases. <br>
<b>Alternative Hypothesis </b> H1: Having private health insurance influence decreases in covid cases.

```python
var1 = enr_data['Private health insurance']
var2 = enr_data['new_cases']
stats.ttest_ind(var2, var1, alternative='two-sided')
```

We reject the null hypothesis and accept the alternative hypothesis since the p-value is less than 0.05 (95 percent significance level). p/2 = 5.4557793987989905e-06/2 < 0.05 which implies our actual hypothesis is true. 

3.
<b>Null Hypothesis H0</b>: Work from home does not influence increase in cases. <br>
<b>Alternative Hypothesis </b> H1: Work from home influence increase in cases.

```python
var1 = enr_data['Work from home']
var2 = enr_data['new_cases']
stats.ttest_ind(var2, var1, alternative='two-sided')
```

We reject the null hypothesis and accept the alternative hypothesis since the p-value is less than 0.05 (95 percent significance level). p/2 = 1.4323274138795011e-05/2 < 0.05 which implies our actual hypothesis is true.
